```python
from tkinter import *
from tkinter.ttk import *

exp=""
def press(n):
    global exp
    exp=exp+str(n)
    equa.set(exp)
def equal():
#to handle error like zero division error
    try:
        global exp
        total=str(eval(exp))
        equa.set(total)
        exp=""
    except:
        equa.set("error")
        exp = "" 
def clear(): 
# Function to clear the contents 
# of text entry box 
    global exp 
    exp = "" 
    equa.set("") 
        
obj=Tk()
obj.title("calculator")
obj.geometry('500x250')   
    
equa = StringVar()  

e1=Entry(obj,textvariable=equa)
e1.grid(columnspan=4, ipadx=70)

btn1=Button(obj,text=0,command=lambda:press(0))
btn1.grid(row=2,column=0)

btn2=Button(obj,text=1,command=lambda:press(1))
btn2.grid(row=2,column=1)

btn3=Button(obj,text=2,command=lambda:press(2))
btn3.grid(row=2,column=2)

btn4=Button(obj,text='+',command=lambda:press('+'))
btn4.grid(row=2,column=3)

btn5=Button(obj,text=3,command=lambda:press(3))
btn5.grid(row=3,column=0)

btn6=Button(obj,text=4,command=lambda:press(4))
btn6.grid(row=3,column=1)

btn7=Button(obj,text=5,command=lambda:press(5))
btn7.grid(row=3,column=2)

btn8=Button(obj,text='-',command=lambda:press('-'))
btn8.grid(row=3,column=3)

btn9=Button(obj,text=6,command=lambda:press(6))
btn9.grid(row=4,column=0)

btn10=Button(obj,text=7,command=lambda:press(7))
btn10.grid(row=4,column=1)

btn11=Button(obj,text=8,command=lambda:press(8))
btn11.grid(row=4,column=2)

btn12=Button(obj,text="/",command=lambda:press('/'))
btn12.grid(row=4,column=3)

btn13=Button(obj,text=9,command=lambda:press(9))
btn13.grid(row=5,column=0)

btn14=Button(obj,text="C",command=lambda:clear())
btn14.grid(row=5,column=1)

btn15=Button(obj,text="=",command=lambda:equal())
btn15.grid(row=5,column=2)

btn16=Button(obj,text="*",command=lambda:press('*'))
btn16.grid(row=5,column=3)

obj.mainloop()
```
